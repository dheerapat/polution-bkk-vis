# pm 2.5 visualization in Bangkok

```python
import pandas as pd
import numpy as np
import datetime as dt
import calmap
import matplotlib.pyplot as plt

from datetime import datetime
```

### read csv data

```python
data = pd.read_csv('pm2020.csv', parse_dates=['Date'])
```

### first 5 row of data

```python
data.head()
```

### make "Date" column be index

```python
data = data.set_index('Date')
```

```python
data.head()
```

### visualization
plot yearly heatmap from '59T' polution station

```python
plt.figure(figsize=(20,10))
calmap.yearplot(data['59T'], fillcolor='lightgrey',daylabels='MTWTFSS',
                dayticks=[0, 2, 4, 6],linewidth=2)
plt.savefig('vis.png')
```
